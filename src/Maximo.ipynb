```python
import findspark
findspark.init()
import pyspark
from pyspark import SparkContext
```

```python
sc = SparkContext("local[*]", "AppMaximo")
```

```python
rdd1 = sc.textFile("/home/alumno/Documents/pg2000.txt")
```

```python
lineas = rdd1.flatMap(lambda linea: linea.split(" "))
```

```python
palmap = lineas.map(lambda pal: (pal, 1))
```

```python
palRed = palmap.reduceByKey(lambda x, y: x + y)
```

```python
palRed.collect()
```

```python
limpio = palRed.filter(lambda x: x[0] != '')
```

```python
cuenta = limpio.reduce(lambda x, y: x if x[1] >= y[1] else y)
```

```python
print("La palabra más repetida es '{}', con {} apariciones".format(cuenta[0], cuenta[1]))
```
